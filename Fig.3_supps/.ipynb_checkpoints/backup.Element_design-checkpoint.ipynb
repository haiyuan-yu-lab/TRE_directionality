# 1. Unidirectional TREs: boundary & active rate 

```python
# We'll send the broad elements for the first batch. If there is a decent number of elements show enhancer activity, we'll send the narrow version of those elements.
```

```python
p = "unidirectional"
d = "distal"
```

```python
bw_pl = PROJECT_DIR_d + "Alignments/brm_C1a_and_C1b_erm_5pl.bw"
bw_mn = PROJECT_DIR_d + "Alignments/brm_C1a_and_C1b_erm_5mn.bw"
bw1 = pyBigWig.open(bw_pl)
bw2 = pyBigWig.open(bw_mn)
```

```python
# Get broad boundary

pause = 60
broads = {}
inputfile = PROJECT_DIR_d + "bed_plot/" + "_".join(["C1", p, d]) + ".bed"
with DelimitedReader(inputfile) as dr:
	for cols in dr:
		chrom, start, end, pstart, pend, tss, _, sign = cols
		center = int(start) + (int(end)-int(start))//2
		if sign == "+":
			bend = int(pend) + pause
			bstart = bend - (bend - center)*2
		else:
			bstart = int(pstart) - pause
			bend = bstart + (center-bstart)*2
	
		reads1 = 0
		reads2 = 0
		pos1 = []
		pos2 = []
		if bw1.intervals(chrom, bstart, bend) != None:
			for zstart, ostop, score in bw1.intervals(chrom, bstart, bend):
				l = min(bend, ostop) - max(bstart, zstart)
				reads1 += abs(score) * l
				pos1.extend([score] * l)
		if bw2.intervals(chrom, bstart, bend) != None:
			for zstart, ostop, score in bw2.intervals(chrom, bstart, bend):
				l = min(bend, ostop) - max(bstart, zstart)
				reads2 += abs(score) * l
				pos2.extend([score] * l)
	
		if sign == "+":
			if reads2 == 0:
				no_reads_on_other_side = True
			else:
				no_reads_on_other_side = False
			if len(pos1) == 1:
				stick = True
			else:
				stick = False
		else:
			if reads1 == 0:
				no_reads_on_other_side = True
			else:
				no_reads_on_other_side = False
			if len(pos2) == 1:
				stick = True
			else:
				stick = False
		broads[(chrom, bstart, bend)] = [pstart, pend, reads1+reads2, no_reads_on_other_side, stick]
```

```python
len(broads)
```

```python
ns = [0,0,0]
broads2 = {}
for k in broads:
	# Filter 1: no reads on the other side
	if broads[k][3]:
		ns[0] += 1
		# Filter 2: keep element size within this range (technical limitation)
		if k[2]-k[1] < 300:
			ns[1] += 1
			# Filter 3: no sticks
			if not broads[k][-1]:
				ns[2] += 1
				broads2[k] = broads[k]
```

```python
ns
```

```python
# Check total counts distribution

tots = [broads2[k][2] for k in broads2]
plt.hist(tots, bins=50)
plt.xlabel("Total read counts", fontsize=16, fontweight="bold")
plt.ylabel("No. elements", fontsize=16, fontweight="bold")
cutoff = np.percentile(tots, 75)
plt.axvline(cutoff, ls="--", c="grey")
plt.text(x=150, y=200, s="Top 25%: " + str(cutoff))
```

```python
# Set 50 as cutoff
# Add new elements at second round

broads3 = {}
outputfile = PROJECT_DIR_o + "exp/broad_boundary.bed"
with DelimitedWriter(outputfile) as dw:
	for k in broads2:
		if broads2[k][2] >= 50:
			broads3[k] = broads2[k]
			if k not in es:
				dw.write(list(k) + [broads3[k][2]])
```
