```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from commandtemplate.conda import run_template_bash
import itertools
```

```python
PROJECT_DIR_d = "/fs/cbsuhy02/storage/yc2553/yc2553/projects/TRE_directionality/PROcap/"
PROJECT_DIR_r = "/fs/cbsuhy02/storage/yc2553/yc2553/projects/TRE_directionality/resources/"
PROJECT_DIR_o = "/fs/cbsuhy02/storage/yc2553/yc2553/projects/TRE_directionality/output/"
conda_env = "DI_test"
```

# Element counts

## Before filtering

```python
s = "C1"
fnames = {"C1": "brm_C1a_and_C1b_erm"}
ps = ["divergent", "unidirectional"]
ds = ["distal_intergenic", "distal_intragenic", "proximal"]
gs = [f"{p}_{d}" for p, d in itertools.product(ps, ds)]
clsname = " ".join(gs)
```

```python
files = [f"{PROJECT_DIR_d}bed_pd/{s}_{g}.bed" for g in gs]
filestr = " ".join(["{i" + str(idx) + "}" for idx in range(len(files))])
run_template_bash(
	"biodatatools annotate_bed_classes -i " + filestr + " -clsname {clsname} -column 3 -o - | cut -f 1-4 | biodatatools summarize_bed_annotation_counts -i - -name {name} -o {o} -func 'lambda d: d[-1]'",
	conda_env=conda_env,
	**{f"i{idx}": file for idx, file in enumerate(files)},
	clsname=clsname,
	name="No.elements",
	o=f"{PROJECT_DIR_d}bed_plot/{s}_counts_raw.bed"
)
```

```python
def plot_counts(df_count, outputfile):
	fig, ax = plt.subplots(1, 1, figsize=(4,6))
	order = [g for g in gs if "distal" in g] + [g for g in gs if "proximal" in g]
	ticklabels = [g.split("_")[0].capitalize() + " (" + g.split("_")[-1] + ")" if "distal" in g else g.split("_")[0].capitalize() for g in order]
			
	g = sns.barplot(data=df_count, y="Group", x="No.elements",
					order=order, orient="h",
					hue="Group",
					palette=["#fb8072", "#fdb462", "#80b1d3"]*2, 
					ax=ax)
	n = 0
	for i in g.containers:
		g.bar_label(i, padding=5, fontsize=17)
		for bar in i:
			if n >=3:
				bar.set_alpha(0.5)
			n += 1
	
	g.set_xlabel("No. elements", fontsize=25, fontweight="bold")
	g.set_ylabel("")
	g.set_xticks([])
	g.tick_params(axis="y", labelsize=17, pad=5)
	g.set_yticks(g.get_yticks())
	g.set_yticklabels(ticklabels)
	g.spines.right.set_visible(False)
	g.spines.top.set_visible(False)
	trans = g.get_yaxis_transform()
	g.annotate("Distal", xy=(-1.3, 0.9), xycoords=trans, ha="center", va="top", fontsize=20, fontweight="bold", rotation=90)
	g.annotate("Proximal", xy=(-1.3, 3.6), xycoords=trans, ha="center", va="top", fontsize=20, fontweight="bold", rotation=90)
	g.plot([-1.15,-1.15],[-0.3,3.4], color="k", transform=trans, clip_on=False)
	g.plot([-1.15,-1.15],[3.7,5.5], color="k", transform=trans, clip_on=False)

	plt.savefig(outputfile, bbox_inches = 'tight', dpi=300)
```

```python
df_raw = pd.read_table(f"{PROJECT_DIR_d}bed_plot/{s}_counts_raw.bed")
df_raw.head(2)
```

```python
outputfile = f"{PROJECT_DIR_o}figures/Fig1b.pdf"
plot_counts(df_raw, outputfile)
```

## After filtering

```python
files = [f"{PROJECT_DIR_d}bed_plot/{s}_{g}.bed" for g in gs]
filestr = " ".join(["{i" + str(idx) + "}" for idx in range(len(files))])
run_template_bash(
	"biodatatools annotate_bed_classes -i " + filestr + " -clsname {clsname} -column 3 -o - | cut -f 1-4 | biodatatools summarize_bed_annotation_counts -i - -name {name} -o {o} -func 'lambda d: d[-1]'",
	conda_env=conda_env,
	**{f"i{idx}": file for idx, file in enumerate(files)},
	clsname=clsname,
	name="No.elements",
	o=f"{PROJECT_DIR_d}bed_plot/{s}_counts_filtered.bed"
)
```

```python
df_filtered = pd.read_table(f"{PROJECT_DIR_d}bed_plot/{s}_counts_filtered.bed")
outputfile = f"{PROJECT_DIR_o}supp_figures/SuppFig1b.pdf"
plot_counts(df_filtered, outputfile)
```

# DI distribution

## Calculate DI

```python
files = [f"{PROJECT_DIR_d}bed_plot/{sample}_{g}.bed" for g in gs]
filestr = " ".join(["{i" + str(idx) + "}" for idx in range(len(files))])
run_template_bash(
	"biodatatools annotate_bed_classes -i " + filestr + " -clsname {clsname} -column 3 -o - | cut -f 1-4 | biodatatools annotate_bed_directionality_index -i - -ipl {ipl} -imn {imn} -o {o}",
	conda_env=conda_env,
	**{f"i{idx}": file for idx, file in enumerate(files)},
	ipl=f"{PROJECT_DIR_d}Alignments/{fnames[sample]}_5pl.bw",
	imn=f"{PROJECT_DIR_d}Alignments/{fnames[sample]}_5mn.bw",
	clsname=clsname,
	o=f"{PROJECT_DIR_d}bed_plot/{sample}_DI.bed"
)
```

```python
df = pd.read_table(f"{PROJECT_DIR_d}bed_plot/{sample}_DI.bed", header=None)
df.columns = ["chrom", "start", "end", "group", "DI"]
df.head(2)
```

## Plot

```python
fig, ax = plt.subplots(1, 1, figsize=(5,4))
g = sns.histplot(data=df[df["group"].str.contains("divergent")], x="DI", hue="group", palette=["#fb8072", "#fdb462", "#80b1d3"], ax=ax)
g.legend_.set_title("Divergent", prop={"weight": "bold", "size":13})
for text in g.legend_.get_texts():
	current_text = text.get_text()
	if "distal" in current_text:
		text.set_text(current_text.split("_")[1].capitalize() + " (" + current_text.split("_")[2] + ")")
	else:
		text.set_text(current_text.split("_")[1].capitalize())
	text.set_fontsize(12)
g.set_ylabel("No. elements", fontsize=14, fontweight="bold")
g.set_xlabel("Directionality index", fontsize=14, fontweight="bold")
g.tick_params(labelsize=12, pad=2)
g.spines.right.set_visible(False)
g.spines.top.set_visible(False)
g.set_ylim([0,590])
outputfile = f"{PROJECT_DIR_o}figures/Fig1c.pdf"
plt.savefig(outputfile, bbox_inches = 'tight', dpi=300)
```
