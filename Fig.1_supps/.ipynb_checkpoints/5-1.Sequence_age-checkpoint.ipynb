# 1. Assign sequence age

Syntenic blocks that overlapped at least 6 bp of an enhancer sequence were considered when assigning the enhancer’s age and architecture. We considered enhancers mapping to one syntenic block or several syntenic blocks of the same age as “simple” enhancer architectures, while enhancers overlapping adjacent syntenic blocks of different ages have “complex” enhancer architectures. Given complex enhancers are composed of multiple sequence ages, we assigned complex enhancer age according to the oldest age. Sequences without an assigned age were excluded from this analysis.

It takes extremely long to do the overlap using python codes. Use bedtools 2.30.0 instead. Do the intersection, then filter the length ≥ 6bp.

"-wao": Write the original A and B entries plus the number of base pairs of overlap between the two features. However, A features w/o overlap are also reported with a NULL B feature and overlap = 0.

"-sorted": For very large B files, invoke a “sweeping” algorithm that requires position-sorted (e.g., sort -k1,1 -k2,2n for BED files) input. When using -sorted, memory usage remains low even for very large files.

For more commands, please refer to https://bedtools.readthedocs.io/en/latest/content/tools/intersect.html#usage-and-option-summary.

```python
def intersect(fdir, fa, fb):
	# control: may further refine it later
	outputfile = fdir + fa.split("/")[-1]
	subprocess.run("bedtools intersect -a " + fa + " -b " + fb + " -sorted -wao > " + outputfile, shell=True)
```

```python
s = "C1"
ps = ["divergent", "unidirectional"]
ds = ["distal", "proximal"]
```

```python
#1 TRE/control files 

filenames1 = []
for p in ps:
	for d in ds:
		for k in ["", "_control"]:
			filenames1.append(PROJECT_DIR_d + "PROcap/bed_plot/" + "_".join([s, p, d]) + k + ".bed")
len(filenames1)
```

```python
#2 Syntenic blocks

filenames2 = glob.glob(PROJECT_DIR_d2 + "UCSC/multiz100way/hg38/chr*.bed")
len(filenames2)
```

```python
pwpool = ProcessWrapPool(len(filenames1))

fdir = PROJECT_DIR_o + "age_overlap/"
fb = " ".join(filenames2)
for fa in filenames1:
	pwpool.run(intersect, args=[fdir, fa, fb])
```

**Output format**

- **Only one B file**

column 1-N: original A entries

column (N+1)-M: original B entries

column (M+1): the length of overlap (bp) between the two features

- **Multiple B files**

column 1-N: original A entries

column (N+1): the file number from which the overlap came

column (N+2)-M: original B entries

column (M+1): the length of overlap (bp) between the two features
