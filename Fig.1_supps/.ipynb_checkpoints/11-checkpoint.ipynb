```python
import pyBigWig
import pandas as pd
import matplotlib.pyplot as plt
import pybedtools
import numpy as np
import seaborn as sns
```

# 1. No. TREs

```python
s = "C1"
ps = ["divergent", "unidirectional"]
ds = ["distal", "proximal"]
ds2 = ["distal_intergenic", "distal_intragenic", "proximal"]
ks = ["pl", "mn"]
ts = ["intergenic", "intragenic"]
```

## 1.1 Raw elements

```python
# Separate distal elements into intergenic and intragenic

def inter_vs_intra(dir):
	d = "distal"
	b1 = pybedtools.BedTool(PROJECT_DIR_r + "Gencode/Human/gencode.v37.annotation.transcripts.union.bed.bgz")
	for p in ps:
		b2 = pybedtools.BedTool(PROJECT_DIR_d + dir + "/" + "_".join([s, p, d]) + ".bed")
		outputfile = PROJECT_DIR_d + dir + "/" + "_".join([s, p, d, "intergenic"]) + ".bed"
		b2.intersect(b1, v=True).saveas(outputfile)
		outputfile = PROJECT_DIR_d + dir + "/" + "_".join([s, p, d, "intragenic"]) + ".bed"
		b2.intersect(b1, u=True).saveas(outputfile)
	results = []
	for p in ps:
		for d in ds2:
			b = pybedtools.BedTool(PROJECT_DIR_d + dir + "/" + "_".join([s, p, d]) + ".bed")
			results.append([(p,d), len(b)])
	df_num = pd.DataFrame(results, columns=["Group", "No.elements"])
	return df_num
```

```python
df_num = inter_vs_intra("bed_pd")
df_num.head(2)
```

```python
def plot_numbers(df_num, outputfile):
	fig, ax = plt.subplots(1, 1, figsize=(4,6))
	order = []
	ticklabels = []
	for d in ds:
		if d == "proximal":
			order.extend([(p,d) for p in ps])
			ticklabels.extend([p.capitalize() for p in ps])
		else:
			for p in ps:
				order.extend([(p,d + "_" + t) for t in ts])
				ticklabels.extend([p.capitalize() + " (" + t + ")" for t in ts])
	g = sns.barplot(data=df_num, y="Group", x="No.elements",
					order=order, orient="h",
					palette=["#fb8072", "#fdb462", "#fb8072", "#fdb462", "#80b1d3", "#80b1d3"], 
					ax=ax)
	n = 0
	for i in g.containers:
		g.bar_label(i, padding=5, fontsize=17)
		for bar in i:
			if n in [2,3,5]:
				bar.set_alpha(0.5)
			n += 1
	
	g.set_xlabel("No. elements", fontsize=25, fontweight="bold")
	g.set_ylabel("")
	g.tick_params(axis="y", labelsize=17, pad=5)
	g.set_xticks([])
	g.set_yticklabels(ticklabels)
	g.spines.right.set_visible(False)
	g.spines.top.set_visible(False)
	trans = g.get_yaxis_transform()
	g.annotate("Distal", xy=(-1.3, 0.9), xycoords=trans, ha="center", va="top", fontsize=20, fontweight="bold", rotation=90)
	g.annotate("Proximal", xy=(-1.3, 3.6), xycoords=trans, ha="center", va="top", fontsize=20, fontweight="bold", rotation=90)
	g.plot([-1.15,-1.15],[-0.3,3.4], color="k", transform=trans, clip_on=False)
	g.plot([-1.15,-1.15],[3.7,5.5], color="k", transform=trans, clip_on=False)

	plt.savefig(outputfile, bbox_inches = 'tight', dpi=300)
```

```python
outputfile = PROJECT_DIR_o + "figures/Fig1B.png"
plot_numbers(df_num, outputfile)
```

# 2. DI distribution

```python
bws = [PROJECT_DIR_d + "Alignments/brm_C1a_and_C1b_erm_5" + k + ".bw" for k in ks]
bw1, bw2 = [pyBigWig.open(bw) for bw in bws]
```

```python
results = []
for p in ps:
	for d in ds2:
		inputfile = PROJECT_DIR_d + "bed_plot/" + "_".join([s, p, d]) + ".bed"
		b = pybedtools.BedTool(inputfile)
		for i in b:
			chrom, start, end = i.fields[:3]
			reads1 = 0
			reads2 = 0
			if bw1.intervals(chrom, int(start), int(end)) != None:
				for zstart, ostop, score in bw1.intervals(chrom, int(start), int(end)):
					l = min(int(end), ostop) - max(int(start), zstart)
					reads1 += abs(score) * l
			if bw2.intervals(chrom, int(start), int(end)) != None:
				for zstart, ostop, score in bw2.intervals(chrom, int(start), int(end)):
					l = min(int(end), ostop) - max(int(start), zstart)
					reads2 += abs(score) * l
			if len(d.split("_")) == 2:
				d2 = d.split("_")[0].capitalize() + " " + d.split("_")[1]
			else:
				d2 = d.capitalize()
			results.append([d2, p, (reads1-reads2)/(reads1+reads2)])
df = pd.DataFrame(results, columns=["Group", "Group2", "DI"])
df.head()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5,4))
g = sns.histplot(data=df[df["Group2"]=="divergent"], x="DI", hue="Group", palette=["#fb8072", "#fdb462", "#80b1d3"], ax=ax)
g.legend_.set_title("Divergent", prop={"weight": "bold", "size":13})
for text in g.legend_.get_texts():
    text.set_fontsize(12)
g.set_ylabel("No. elements", fontsize=14, fontweight="bold")
g.set_xlabel("Directionality index", fontsize=14, fontweight="bold")
g.tick_params(labelsize=12, pad=2)
g.spines.right.set_visible(False)
g.spines.top.set_visible(False)
g.set_ylim([0,590])
outputfile = PROJECT_DIR_o + "figures/Fig1C.png"
plt.savefig(outputfile, bbox_inches = 'tight', dpi=300)
```
