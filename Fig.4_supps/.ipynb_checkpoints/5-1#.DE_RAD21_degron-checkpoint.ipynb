```python
index = ["U1", "U2", "T1", "T2"]
results = [["U"],
		   ["U"],
		   ["T"],
		   ["T"]
		  ]
metadata = pd.DataFrame(results, columns=["Condition"], index=index)
metadata.head()
```

## 1.2 Count data

```python
s = "HCT116"
ks = ["pl", "mn"]
ts = ["maximum", "minimum"]
labels = ["Up", "Unchanged", "Down"]
ps = ["divergent", "unidirectional"]
ds = ["proximal", "distal"]
gs = [p+"_"+d for p in ps for d in ds]
```

```python
reps = list(metadata.index)
bws = {}
for r in reps:
	bws[r] = [PROJECT_DIR_d + "Alignments/" + "_".join(["RAD21", r, "5"+k]) + ".bw" for k in ks]
```

```python
def get_read_matrix(outputfile):
	results = []
	index = []
	for p in ps:
		for d in ds:
			inputfile = PROJECT_DIR_d + "bed_plot/" + "_".join([s, p, d]) + ".bed"
			with DelimitedReader(inputfile) as dr:
				for cols in dr:
					chrom, start, end = cols[:3]
					start = int(start)
					end = int(end)
					sign = cols[-1]
	
					for t in ts:
						index.append("_".join([chrom, str(start), str(end), t, p, d]))
						row = []
						for r in bws:
							pl, mn = bws[r]
							bw1 = pyBigWig.open(pl)
							bw2 = pyBigWig.open(mn)
							reads1 = 0
							reads2 = 0
							
							if bw1.intervals(chrom, start, end+1) != None:
								for start2, end2, score in bw1.intervals(chrom, start, end+1):
									l = min(end+1, end2) - max(start, start2)
									reads1 += abs(score) * l
							if bw2.intervals(chrom, start, end+1) != None:
								for start2, end2, score in bw2.intervals(chrom, start, end+1):
									l = min(end+1, end2) - max(start, start2)
									reads2 += abs(score) * l
							
							if t == "maximum":
								row.append(reads1 if sign == "+" else reads2)
							else:
								row.append(reads2 if sign == "+" else reads1)
						results.append(row)
				
	df = pd.DataFrame(results, columns=list(bws.keys()), index=index)
	df = df.T
	df.to_csv(outputfile, sep="\t")
```

# 2. DE analysis

## 2.1 Run DESeq2

```python
# https://www.bioconductor.org/packages/release/bioc/vignettes/DESeq2/inst/doc/DESeq2.html#indfilttheory
# Pre-filtering step can be omitted entirely and just rely on the independent filtering procedures
```

```python
def DE_analysis(counts_df, outputfile):	
	# Create DeseqDataSet object
	dds = DeseqDataSet(
	    counts=counts_df,
	    metadata=metadata,
	    design_factors="Condition",
		ref_level=["Condition","U"],
		n_cpus=8
	)
	# Fit dispersions and LFCs
	dds.deseq2()
	# Compute p-values and adjusted p-values
	stat_res = DeseqStats(dds, n_cpus=8)
	stat_res.summary()
	# Get results
	res = stat_res.results_df
	res["log2baseMean"] = np.log2(res["baseMean"])
	conditions = [
	    (res['padj'] < 0.05) & (res['log2FoldChange'] > 1),
	    (res['padj'] < 0.05) & (res['log2FoldChange'] < -1),
	]
	choices = ['Up', 'Down']
	res['Label'] = np.select(conditions, choices, default='Unchanged')
	res.to_csv(outputfile, sep="\t")
	return res
```

```python
inputfile = PROJECT_DIR_o + "DE/" + "_".join([s, "counts"]) + ".txt"
counts_df = pd.read_table(inputfile, index_col=0)
outputfile = PROJECT_DIR_o + "DE/" + "_".join([s, "DE"]) + ".txt"
res = DE_analysis(counts_df, outputfile)
```

## 2.2 MA plot

```python
def MA_plot(res):
	g = sns.scatterplot(data=res, x="log2baseMean", y="log2FoldChange", 
				hue="Label", hue_order=["Up", "Unchanged", "Down"],
				palette=["#fb8072","#d9d9d9","#80b1d3"],
				alpha=0.5, s=10)
	g.spines.right.set_visible(False)
	g.spines.top.set_visible(False)
	g.set_xlabel(r"$\bf{log}_2(mean\ expression)$", fontsize=16, fontweight="bold")
	g.set_ylabel(r"$\bf{log}_2(6h/0h)$", fontsize=16, fontweight="bold")
	g.legend(loc="lower right", title=None, fontsize=12)
```

```python
MA_plot(res)
```

## 2.3 Volcano plot

```python
def volcano_plot(res, d, t, ax, title):
	index = [e for e in list(res.index) if d in e and t in e]
	res2 = res.loc[index]
	res2["-log10padj"] = -np.log10(res2["padj"])
	g = sns.scatterplot(data=res2, x="log2FoldChange", y="-log10padj", 
				hue="Label", hue_order=labels,
				palette=["#fb8072","#d9d9d9","#80b1d3"], 
				alpha=0.5, s=10, ax=ax)
	g.legend(title=None, fontsize=12)
	g.spines.right.set_visible(False)
	g.spines.top.set_visible(False)
	g.set_title(title, fontsize=20, fontweight="bold")
	g.set_yticks([0,10,20,30,40])
	g.set_ylim([-5,48])
	g.set_ylabel(r"$-\bf{log}_{10}(q-value)$", fontsize=16, fontweight="bold")
	g.set_xlabel(r"$\bf{log}_2(6h/0h)$", fontsize=16, fontweight="bold")
	
	g.legend(loc="upper left")
	legend = g.legend()
	for label in labels:
		legend.get_texts()[labels.index(label)].set_text(label+" (n=" + str(len(res2[res2["Label"]==label])) + ")")

	for text in legend.get_texts():
	    text.set_fontsize(14)
		
	g.set_xlim([-10.5, 10.5])
```

```python
for d in ds:
	for t in ts:
		fig, ax = plt.subplots(1,1,figsize=(6,4))
		volcano_plot(res, d, t, ax, t.capitalize()+" TSS")
```

# 3. Maximum vs. minimum

```python
effects = defaultdict(dict)
for t in ts:
	for d in ds:
		index = [e for e in list(res.index) if d in e and t in e]
		res2 = res.loc[index]
		for label in labels:
			effects[(t,d)][label] = set(["_".join(e.split("_")[:3]) for e in res2[res2["Label"]==label].index])
```

```python
dfs = {}
for d in ds:
	results = []
	for label in labels:
		row = [round(len(effects[("maximum",d)][label2] & effects[("minimum",d)][label])/len(effects[("minimum",d)][label])*100,1) for label2 in labels[:-1]]
		row.append(round(100-sum(row),1))
		row = [str(v)+"%" for v in row]
		results.append(row)
	dfs[d] = pd.DataFrame(results, columns=labels, index=labels)
```
