# 1. Unidirectional TREs: boundary & active rate 

```python
# We'll send the broad elements for the first batch. If there is a decent number of elements show enhancer activity, we'll send the narrow version of those elements.
```

```python
p = "unidirectional"
d = "distal"
```

```python
bw_pl = PROJECT_DIR_d + "PROcap/bw/C1_5_end_pl.bw"
bw_mn = PROJECT_DIR_d + "PROcap/bw/C1_5_end_mn.bw"
bw1 = pyBigWig.open(bw_pl)
bw2 = pyBigWig.open(bw_mn)
```

```python
# Get broad boundary

pause = 60
broads = {}
inputfile = PROJECT_DIR_d + "PROcap/bed_plot/" + "_".join(["C1", p, d]) + ".bed"
with DelimitedReader(inputfile) as dr:
	for cols in dr:
		chrom, start, end, pstart, pend, tss, _, sign = cols
		center = int(start) + (int(end)-int(start))//2
		if sign == "+":
			bend = int(pend) + pause
			bstart = bend - (bend - center)*2
		else:
			bstart = int(pstart) - pause
			bend = bstart + (center-bstart)*2
	
		reads1 = 0
		reads2 = 0
		pos1 = []
		pos2 = []
		if bw1.intervals(chrom, bstart, bend) != None:
			for zstart, ostop, score in bw1.intervals(chrom, bstart, bend):
				l = min(bend, ostop) - max(bstart, zstart)
				reads1 += abs(score) * l
				pos1.extend([score] * l)
		if bw2.intervals(chrom, bstart, bend) != None:
			for zstart, ostop, score in bw2.intervals(chrom, bstart, bend):
				l = min(bend, ostop) - max(bstart, zstart)
				reads2 += abs(score) * l
				pos2.extend([score] * l)
	
		if sign == "+":
			if reads2 == 0:
				no_reads_on_other_side = True
			else:
				no_reads_on_other_side = False
			if len(pos1) == 1:
				stick = True
			else:
				stick = False
		else:
			if reads1 == 0:
				no_reads_on_other_side = True
			else:
				no_reads_on_other_side = False
			if len(pos2) == 1:
				stick = True
			else:
				stick = False
		broads[(chrom, bstart, bend)] = [pstart, pend, reads1+reads2, no_reads_on_other_side, stick]
```

```python
len(broads)
```

```python
ns = [0,0,0]
broads2 = {}
for k in broads:
	# # Filter 1: no reads on the other side
	if broads[k][3]:
		ns[0] += 1
		# Filter 2: keep element size within this range (technical limitation)
		if k[2]-k[1] < 320:
			ns[1] += 1
			# Filter 3: no sticks
			if not broads[k][-1]:
				ns[2] += 1
				broads2[k] = broads[k]
```

```python
ns
```

```python
# Check total counts distribution

tots = [broads2[k][2] for k in broads2]
plt.hist(tots, bins=50)
plt.xlabel("Total read counts", fontsize=16, fontweight="bold")
plt.ylabel("No. elements", fontsize=16, fontweight="bold")
cutoff = np.percentile(tots, 75)
plt.axvline(cutoff, ls="--", c="grey")
plt.text(x=150, y=200, s="Top 25%: " + str(cutoff))
```

```python
# Round 1

inputfile = PROJECT_DIR_o + "exp/broad_boundary_manual.bed"
b = pybedtools.BedTool(inputfile)
es = [(i.fields[0], int(i.fields[1]), int(i.fields[2])) for i in b]
```

```python
# Set 50 as cutoff
# Add new elements at second round

broads3 = {}
outputfile = PROJECT_DIR_o + "exp/broad_boundary.bed"
with DelimitedWriter(outputfile) as dw:
	for k in broads2:
		if broads2[k][2] >= 50:
			broads3[k] = broads2[k]
			if k not in es:
				dw.write(list(k) + [broads3[k][2]])
```

```python
args = {}
for s in bws:
	args[s] = pm.as_list(pm.open_pyBigWig(bws[s][0]), 
								  pm.open_pyBigWig(bws[s][1]))
# Use peak calls from merged sample
args["Broad_boundary"] = BED3Reader.read_all(genomictools.GenomicCollection, 
									PROJECT_DIR_o + "exp/broad_boundary.bed")
args["DNase"] = pm.open_pyBigWig(PROJECT_DIR_d + "ENCODE/ENCFF414OGC.bigWig")
args["H3K27ac"] = pm.open_pyBigWig(PROJECT_DIR_d + "ENCODE/ENCFF849TDM.bigWig")

args = pm.as_dict(list(args.keys()), pm.as_list(*args.values()))

procap_formatter = {"asym_pos_neg":True, "plot_kw":{"width":2}}
dnase_formatter = {"plot_kw":{"width":1, "color":"#b3de69"}}
h3k27ac_formatter = {"plot_kw":{"width":1, "color":"#FFD700"}}
formatters={"DNase": dnase_formatter,
			"H3K27ac": h3k27ac_formatter,
		   }
height_ratios = {}
for k in args:
	if "PROcap" in k:
		formatters[k] = procap_formatter
		height_ratios[k] = 8
	elif "peaks" in k:
		height_ratios[k] = 3
	else:
		height_ratios[k] = 5

plot_genome_view_kw = {"args": args,
					   "formatters": formatters,
					   "height_ratios": height_ratios,
					   "group_autoscales":[[k for k in list(args.keys()) if "5" in k], [k for k in list(args.keys()) if "3" in k]]
					  }
plot_genome_view_kw = pm.as_dict(list(plot_genome_view_kw.keys()), pm.as_list(*plot_genome_view_kw.values()))
```

```python
def plot_examples(all_items, n, n_subplots, plot_genome_view_kw):
	if (n+1)*n_subplots > len(all_items):
		selected = all_items[n*n_subplots:]
	else:
		selected = all_items[n*n_subplots:(n+1)*n_subplots]
	pm.plot_regions_genome_views(genomictools.GenomicCollection(regions), 
								 output=outputfile,
								 expand_size=600,
								 selected=selected,
								 plot_genome_view_kw=plot_genome_view_kw,
								 plt_figure_change_properties_kw={"locator_prop":{"nbins":3, "axis":"x"}},
								 compose_SVG_panel_kw={"w":500, "cols":5, "save_fig_kw":{"bbox_inches":"tight"}}
							)
	plt.close()
```

```python
n_plot = 1
n_subplots = 35

regions = BED3Reader.read_all(genomictools.GenomicCollection, PROJECT_DIR_o + "exp/broad_boundary.bed")
all_items = [i for i in range(len(regions))]
for n in range(n_plot):
	outputfile = PROJECT_DIR_o + "exp/elements_" + str(n+1) + ".pdf"
	plot_examples(all_items, n, n_subplots, plot_genome_view_kw)
```
