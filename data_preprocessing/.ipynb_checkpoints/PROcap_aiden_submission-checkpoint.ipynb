## Calculate CoPRO RNA len

```python
from biodata.bed import BEDGraphReader
import itertools
import numpy as np
```

```python
rnalens = {}
for lib_name in seq_lib_names:
	print(lib_name)
	if lib_name in rnalens:
		continue
	with BEDGraphReader(f'/local/storage/kl945/PROcap/Alignments/{lib_name}_dpl.bed.gz', dataValueType=lambda s: list(map(int, s.split(',')))) as br:
		dpl_len = list(itertools.chain.from_iterable([b.dataValue for b in br]))
	with BEDGraphReader(f'/local/storage/kl945/PROcap/Alignments/{lib_name}_dmn.bed.gz', dataValueType=lambda s: list(map(int, s.split(',')))) as br:
		dmn_len = list(itertools.chain.from_iterable([b.dataValue for b in br]))
	rnalens[lib_name] = np.median(dpl_len + dmn_len)
```

### Raw fastq

```python
from biodata.baseio import BaseReader
def get_read_length(f):
	with BaseReader("/local/storage/kl945/PROcap/" + f) as br:
		br.read()
		s1 = br.read()
		br.read()
		s2 = br.read()
		if len(s1) != len(s2):
			raise Exception()
		return len(s1)
	
	
header=["lab", "award", "dataset", "file_format", "run_type", "platform", "output_type", "replicate", "submitted_file_name", "paired_end", "paired_with", "aliases", "read_length"]
contents = []
for lib_name in lib_names:
	entry = aiden_procap_info.loc[aiden_procap_info["Library Name"] == lib_name].to_dict("records")[0]
	sample_id = entry['Sample Identifier']
	biological_replicate_number = entry["Biological Replicate"]
	technical_replicate_number = entry["Technical Replicate"]
	dataset=f"haiyuan-yu:procap-experiment-{batch_prefix}-{sample_id}"
	file_format = "fastq"
	run_type = "paired-ended"
	platform = "encode:NovaSeq6000"
	output_type = "reads"
	replicate = f"haiyuan-yu:procap-replicate-{batch_prefix}-{sample_id}-{biological_replicate_number}_{technical_replicate_number}"
	
	for row in procap_info.loc[procap_info["Library Name"] == lib_name].iterrows():
		
		r1, r2 = "RawData/" + row[1]["Sequencing Read 1"], "RawData/" + row[1]["Sequencing Read 2"]
		
		seq_lib_name = row[1]["Sequenced Library Name"]
		if seq_lib_name not in seq_lib_names:
			continue
		if seq_lib_name.endswith("reseq2"):
			suffix = f"{sample_id}-{biological_replicate_number}_{technical_replicate_number}-reseq2"
		else:
			suffix = f"{sample_id}-{biological_replicate_number}_{technical_replicate_number}"
	
		submitted_file_name = r1
		paired_end = "1"
		paired_with = ""
		aliases = [f"haiyuan-yu:procap-read1-{batch_prefix}-{suffix}"]
		read_length = get_read_length(r1)
		contents.append([lab, award, dataset, file_format, run_type, platform, output_type, replicate, submitted_file_name, paired_end, paired_with, aliases, read_length])

		submitted_file_name = r2 
		paired_end = "2"
		paired_with = f"haiyuan-yu:procap-read1-{batch_prefix}-{suffix}"
		aliases = [f"haiyuan-yu:procap-read2-{batch_prefix}-{suffix}"]
		read_length = get_read_length(r2)
		contents.append([lab, award, dataset, file_format, run_type, platform, output_type, replicate, submitted_file_name, paired_end, paired_with, aliases, read_length])
# 	del dataset, file_format, run_type, platform, output_type, replicate, submitted_file_name, paired_end, paired_with, aliases, read_length

contents.sort()
df = pd.DataFrame(contents, columns=header)
# df.to_csv(encode_submission_folder + "submit_file_fastq.txt", sep="\t", index=False)
df
```

```python
header=["lab", "award", "dataset", "file_format", "file_format_type", "output_type", "submitted_file_name", "aliases", "derived_from", "assembly", "step_run"]
contents = []

for sample_id in sample_ids:
	dataset=f"haiyuan-yu:procap-experiment-{batch_prefix}-{sample_id}"
	file_format = "bed"
	step_run = "haiyuan-yu:pro-cap-run-step-4"
	file_format_type = "bed3+"
	assembly = "GRCh38"
	suffice = []
	slibs = []
	for r in aiden_procap_info.loc[aiden_procap_info["Sample Identifier"] == sample_id].to_dict("records"):
		sample_id = r["Sample Identifier"]
		lib_name = r["Library Name"]
		biological_replicate_number = r["Biological Replicate"]
		technical_replicate_number = r["Technical Replicate"]
		for seq_lib_name in procap_info.loc[procap_info["Library Name"] == lib_name]["Sequenced Library Name"]:
			slibs.append(seq_lib_name)
			if seq_lib_name.endswith("reseq2"):
				suffice.append(f"{sample_id}-{biological_replicate_number}_{technical_replicate_number}-reseq2")
			else:
				suffice.append(f"{sample_id}-{biological_replicate_number}_{technical_replicate_number}")
				
	derived_from = ([f"haiyuan-yu:procap-bigwig-plus-{batch_prefix}-{suffix}" for suffix in suffice]
					+ [f"haiyuan-yu:procap-bigwig-minus-{batch_prefix}-{suffix}" for suffix in suffice])
	output_type = "bidirectional peaks"
	output_type_str = "bidirectional_peaks"
	aliases = [f"haiyuan-yu:procap-{output_type_str}-{batch_prefix}-{sample_id}"]
	submitted_file_name=f"SubmissionFolder/alignmentmerge_{'_and_'.join(slibs)}_1_bidirectional_peaks.bed.gz"
	contents.append([lab, award, dataset, file_format, file_format_type, output_type, submitted_file_name, aliases, derived_from, assembly, step_run])
	
	
	
	output_type = "unidirectional peaks"
	output_type_str = "unidirectional_peaks"
	aliases = [f"haiyuan-yu:procap-{output_type_str}-{batch_prefix}-{sample_id}"]
	submitted_file_name=f"SubmissionFolder/alignmentmerge_{'_and_'.join(slibs)}_1_unidirectional_peaks.bed.gz"
	contents.append([lab, award, dataset, file_format, file_format_type, output_type, submitted_file_name, aliases, derived_from, assembly, step_run])
	
	del dataset, file_format, file_format_type, output_type, submitted_file_name, aliases, derived_from, assembly, step_run
contents.sort()
df = pd.DataFrame(contents, columns=header)
# df.to_csv(encode_submission_folder + "submit_file_bed.txt", sep="\t", index=False)
df
```
